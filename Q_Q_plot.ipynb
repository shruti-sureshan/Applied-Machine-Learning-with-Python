# Q-Q Plot

```python
#Q-Q plot
import numpy as np 
import pylab 
import scipy.stats as stats

# N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)  #loc is mean and scale is std dev  , we generate 1000 samples

# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))


```

```python
import matplotlib.pyplot as plt
```

```python
# generate 100 sanples from N(20,5)
measurements = np.random.normal(loc = 0, scale = 1, size=100) 
plt.xlim(-4, 4)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
# generate 10000 sanples from N(20,5)
measurements = np.random.normal(loc = 0, scale = 1, size=10000) 

plt.xlim(-1,1)

stats.probplot(measurements, dist="norm", plot=pylab) #compare with normal dist
pylab.show()
```

```python
# generate 10000 sanples from N(20,5)
measurements = np.random.normal(loc = 0, scale = 1, size=10000) 
plt.xlim(-4, 4)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
# generate 100 sanples from N(20,5)
measurements = np.random.uniform(low=-1, high = 1, size=100) 
plt.xlim(-4, 4)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```
